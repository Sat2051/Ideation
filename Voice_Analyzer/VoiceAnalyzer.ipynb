```python
import syslog

from scipy.io.wavfile import read
import numpy as np
import matplotlib.pyplot as plt


(fs, x) = read('self_001.wav')

print(fs) # sampling rate
syslog.syslog(200,'logging x value')
print(x)  # read the file


print(x.size)  # checking the length of the array

# finding how many seconds that sound is


print(x.size/fs)


plt.plot(x)
plt.show()

t = np.arange(x.size)/fs

plt.plot(t,x)
plt.show()

# taking a sample between the array x





```
